```python
from minio import Minio
import os
from dotenv import load_dotenv
from datetime import datetime, timedelta
import time
import requests
from minio.error import S3Error
from io import BytesIO
```

```python
load_dotenv()

access_key = os.getenv("MINIO_ACCESS_KEY")
secret_key = os.getenv("MINIO_SECRET_KEY")
```

```python
client = Minio(
    endpoint="localhost:9000",
    access_key=access_key,
    secret_key=secret_key,
    secure=False
)

backet_html ="raw-html"

if not client.bucket_exists(backet_html):
    client.make_bucket(backet_html)
```

```python
print("Total backets: ", len(client.list_buckets()))
print(client.list_buckets())
```

```python
def minio_put_object(client, bucket_name, object_name, url, html_content):
    try:
        client.put_object(
            bucket_name,
            object_name,
            data=BytesIO(html_content),
            length=len(html_content),
            content_type="text/html"
        )
        print(f"Страница {url} успешно загружена как '{object_name}'")
    except S3Error as err:
        print("Ошибка загрузки:", err)
```

```python
start_year = 2024
end_year = 2024


for year in range(start_year, end_year - 1, -1):
    start_date = datetime(year, 12, 5)
    end_date = datetime(year, 4, 1)

    current_date = start_date
    while current_date >= end_date:
        date_str = current_date.strftime("%Y%m%d")  # Формат YYYYMMDD
        print(date_str)
        
        url = f"https://vrnfso.ru/download/{year}/{date_str}_rez.htm"

        # Проверяем, существует ли файл по этому URL
        response = requests.head(url)
        if response.status_code == 200:
            print(f"Доступен: {url}")

            page = requests.get(url)
            html_content = page.content
            minio_put_object(client, backet_html, date_str, url, html_content)
            
        time.sleep(5)
        # Переход к предыдущему дню
        current_date -= timedelta(days=1)
```
